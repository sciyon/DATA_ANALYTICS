**TASK: Take 10 moderately sized subsamples your dataset and create a correlation plot for each subsample**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Convert data to pandas DataFrame
data = pd.DataFrame(mice_protein_expression.data['features'])
data = data.rename(columns=lambda x: x.split('_')[0]) # Rename columns to remove '_N' suffix
data['class'] = mice_protein_expression.data['targets'] # Add class column

# Take 10 subsamples of size 100
subsamples = [data.sample(100, random_state=i) for i in range(10)]

# Create correlation plots for each subsample with top 10 most correlated features
for i, subsample in enumerate(subsamples):
    corr = subsample.corr().abs().unstack().sort_values(kind="quicksort", ascending=False).reset_index()
    corr = corr[corr['level_0'] != corr['level_1']]
    top_corr_features = corr.loc[:, 'level_0'].tolist()[:10] + corr.loc[:, 'level_1'].tolist()[:10]
    top_corr_features = list(set(top_corr_features))
    
    plt.figure(figsize=(10, 8))
    sns.heatmap(subsample[top_corr_features].corr(), annot=True, cmap='coolwarm')
    plt.title(f'Correlation Plot for Subsample {i+1} (Top 10 Features)')
    plt.show()
```

**TASK: Take 10 moderately sized subsamples and create a pairplot of the feature variable in relation to the target variable**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from ucimlrepo import fetch_ucirepo

# Fetch the dataset
mice_protein_expression = fetch_ucirepo(id=342)

# Separate features and targets
X = mice_protein_expression.data.features
y = mice_protein_expression.data.targets

# Assuming y is a single column DataFrame, convert it to a Series for easier handling
if isinstance(y, pd.DataFrame):
    y = y.iloc[:, 0]

# Function to generate a subsample and plot its pairplot
def plot_subsample_pairplot(subsample_size):
    # Generate a random index for the subsample
    indices = np.random.choice(X.index, subsample_size, replace=False)
    
    # Select the subsample
    X_subsample = X.loc[indices]
    y_subsample = y.loc[indices]
    
    # Plot the pairplot
    plt.figure(figsize=(12, 8))
    sns.pairplot(X_subsample, hue=y_subsample.name, vars=X.columns, diag_kind='kde')
    plt.title(f'Subsample Pairplot (Size: {subsample_size})')
    plt.show()

# Generate and plot 10 moderately sized subsamples
for i in range(1, 11):
    plot_subsample_pairplot(int(len(X)*0.1))  # Adjust the multiplier as needed to control the subsample size
```

**TASK: Perform SVM modelling and check the performance of your model, improve the performance of your model using grdisearch**

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Convert data to pandas DataFrame
data = pd.DataFrame(mice_protein_expression.data['features'])
data = data.rename(columns=lambda x: x.split('_')[0]) # Rename columns to remove '_N' suffix
data['class'] = mice_protein_expression.data['targets'] # Add class column

# Encode class labels
le = LabelEncoder()
data['class'] = le.fit_transform(data['class'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import GridSearchCV
from sklearn.impute import SimpleImputer
import pandas as pd

# Assuming 'data' is your DataFrame and 'class' is the column you want to predict
# Split data into features and target
X = data.drop('class', axis=1)
y = data['class']

# Check for missing values
if X.isnull().sum().any():
    # Use SimpleImputer to fill missing values with the mean of the column
    imputer = SimpleImputer(strategy='mean')
    X_imputed = imputer.fit_transform(X)
else:
    X_imputed = X.values  # No missing values, proceed normally

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y, test_size=0.2, random_state=42)

# Train the SVM model
svm = SVC()
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)

# Evaluate the model
print('Basic SVM Accuracy:', accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid for GridSearchCV
param_grid = {
    'C': [0.1, 1, 10, 100],
}

# Initialize the GridSearchCV object
grid_search = GridSearchCV(SVC(), param_grid, cv=5, scoring='accuracy')

# Fit the GridSearchCV object to the training data
grid_search.fit(X_train, y_train)

# Get the best parameters
best_params = grid_search.best_params_

# Print the best parameters
print("Best Parameters:", best_params)

# Use the best parameters to retrain the model
best_svm = SVC(C=best_params['C'])
best_svm.fit(X_train, y_train)

# Predict on the test set
y_pred_best = best_svm.predict(X_test)

# Evaluate the model with the best parameters
print('Improved SVM Accuracy:', accuracy_score(y_test, y_pred_best))
print(classification_report(y_test, y_pred_best))
```
