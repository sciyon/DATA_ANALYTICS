**Is there a relationship between CO and Benzene?**

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Load data with correct column names
names = ['Date', 'Time', 'CO(GT)', 'PT08.S1(CO)', 'NMHC(GT)', 'C6H6(GT)', 'PT08.S2(NMHC)', 'NOx(GT)', 'PT08.S3(NOx)', 'NO2(GT)', 'PT08.S4(NO2)', 'PT08.S5(O3)', 'T', 'RH', 'AH']
data = pd.read_csv('AirQualityUCI.csv', names=names, sep=';')

# Create boolean mask for numerical columns
numerical_mask = data.dtypes != object

# Remove rows with missing CO or Benzene values
data = data[~data['CO(GT)'].isna() & ~data['C6H6(GT)'].isna()]

# Convert numerical columns to float
data.loc[:, numerical_mask] = data.loc[:, numerical_mask].astype(float)

# Extract feature (CO) and target (Benzene)
X = data['CO(GT)'].values.reshape(-1,1)
y = data['C6H6(GT)'].values

# Create linear regression model
model = LinearRegression()

# Fit the model
model.fit(X, y)

# Print coefficients
print(f"Intercept: {model.intercept_}")
print(f"Slope: {model.coef_[0]}")
```
