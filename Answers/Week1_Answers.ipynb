```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
data = pd.read_csv('AirQualityUCI.csv', delimiter=';')

# Calculate the correlation matrix
corr = data.corr(numeric_only=True)

# Create a heatmap of the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Plot')
plt.show()
```

```python
# Visualize the distribution of each column
data.hist(figsize=(15, 10))
plt.suptitle('Distribution of Each Column')
plt.show()

# Determine the probability distribution based on the histograms
# For example, you can use statistical tests like the Kolmogorov-Smirnov test to fit distributions.
```
