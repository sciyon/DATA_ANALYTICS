```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the data into a pandas DataFrame
df = pd.read_csv('AirQualityUCI.csv', sep=';')

# Convert the Date and Time columns to a datetime index with the correct format
df['DateTime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'], format='%d/%m/%Y %H.%M.%S')
df = df.set_index('DateTime')

# Select the gas concentration columns
gas_cols = ['CO(GT)', 'NMHC(GT)', 'C6H6(GT)', 'NOx(GT)', 'NO2(GT)']

# Create the plot
plt.figure(figsize=(12, 6))
for col in gas_cols:
    plt.plot(df.index, df[col], label=col)
plt.xlabel('Date and Time')
plt.ylabel('Gas Concentration')
plt.title('Time Series of Gas Concentrations')
plt.legend()
plt.show()
```

```python
# Ensure the index is of datetime type
assert isinstance(df.index, pd.DatetimeIndex), "Index is not of datetime type"

# Check data types of columns to be plotted
for col in gas_cols:
    assert df[col].dtype in [float, int], f"Column {col} is not of numerical type"

# Plot a single column as a test
plt.figure(figsize=(12, 6))
plt.plot(df.index, df[gas_cols[0]], label=gas_cols[0])
plt.xlabel('Date and Time')
plt.ylabel('Gas Concentration')
plt.title('Time Series of Gas Concentrations')
plt.legend()
plt.show()
```
